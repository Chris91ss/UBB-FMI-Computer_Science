```python
# Exercise 1
x = var('x')
u = function('u')(x)

eq_1 = diff(u, x, 2) + 5 * diff(u, x) - 7 * u == x^2 + 5 * x - 7
sol_1 = desolve(eq_1, u)
show(sol_1)

pol_1 = sol_1.subs(_K1=0, _K2=0).expand()
show(pol_1)
show(plot(pol_1, -10, 5))

pol_1_d1 = diff(pol_1, x)
pol_1_d2 = diff(pol_1, x, 2)
show(pol_1_d1.subs(x=pi/2))
show(pol_1_d2.subs(x=pi/2))
```
