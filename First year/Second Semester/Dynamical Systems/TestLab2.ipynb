```python
# Exercise 5
from sage.plot.line import Line

def f(x):
    return 0.02 * x * (100 - x)

n = 40

x = []
starting_point = 10
xk = starting_point
x.append([starting_point, f(starting_point)])
for k in range(1, n + 1):
    xk = f(xk)
    x.append([xk, f(xk)])
    
line_1 = line(x, marker='o', color='blue')

x = []
starting_point = 80
xk = starting_point
x.append([starting_point, f(starting_point)])
for k in range(1, n + 1):
    xk = f(xk)
    x.append([xk, f(xk)])
    
line_2 = line(x, marker='o', color='red')

x = []
starting_point = 95
xk = starting_point
x.append([starting_point, f(starting_point)])
for k in range(1, n + 1):
    xk = f(xk)
    x.append([xk, f(xk)])
    
line_3 = line(x, marker='o', color='green')

show(line_1 + line_2 + line_3)
```
